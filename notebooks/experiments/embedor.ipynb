```python
%autoreload 2
from src.plotting import *
def gen_dla(
    n_dim=100, n_branch=3, branch_length=1000, rand_multiplier=2, seed=37, sigma=4
):
    np.random.seed(seed)
    M = np.cumsum(-1 + rand_multiplier * np.random.rand(branch_length, n_dim), 0)
    for i in range(n_branch - 1):
        ind = np.random.randint(branch_length)
        new_branch = np.cumsum(
            -1 + rand_multiplier * np.random.rand(branch_length, n_dim), 0
        )
        M = np.concatenate([M, new_branch + M[ind, :]])

    noise = np.random.normal(0, sigma, M.shape)
    M = M + noise

    # returns the group labels for each point to make it easier to visualize
    # embeddings
    C = np.array([i // branch_length for i in range(n_branch * branch_length)])

    return M, C


####### tree #######
X, labels = gen_dla(n_dim=100, n_branch=8, sigma=4, branch_length=500)

embedor = EmbedOR(exp_params)
embedding = embedor.fit_transform(X)
plot_graph_2D(embedding, embedor.G, edge_width=0.05, node_size=0.1)

umap_emb = umap.UMAP(n_neighbors=15, min_dist=0.1, metric='euclidean').fit_transform(X)
plot_graph_2D(umap_emb, embedor.G, edge_width=0.05, node_size=0.1)

tsne_emb = TSNE(n_components=2, perplexity=30, n_iter=300).fit_transform(X)
plot_graph_2D(tsne_emb, embedor.G, edge_width=0.05, node_size=0.1)

phate_emb = phate.PHATE(n_jobs=-2).fit_transform(X)
plot_graph_2D(phate_emb, embedor.G, edge_width=0.05, node_size=0.1)
```
